# Configuration

```python
TRAIN_FILE = './data/train.csv'
TEST_FILE = './data/test.csv'
VERSION = '0.0'
```

# Imports

```python
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
import xgboost as xgb
import lightgbm as lgb
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Load Data and Analyze

```python
df_train = pd.DataFrame.from_csv(TRAIN_FILE)
df_test = pd.DataFrame.from_csv(TEST_FILE)
```

```python
df_train.info()
df_train.head()
```

```python
df_test.info()
df_test.head()
```

# Preprocessing

```python
tokenizer0 = Tokenizer()
tokenizer0.fit_on_texts(df_train.X0.tolist() + df_test.X0.tolist())
tokenizer1 = Tokenizer()
tokenizer1.fit_on_texts(df_train.X1.tolist() + df_test.X1.tolist())
tokenizer2 = Tokenizer()
tokenizer2.fit_on_texts(df_train.X2.tolist() + df_test.X2.tolist())
tokenizer3 = Tokenizer()
tokenizer3.fit_on_texts(df_train.X3.tolist() + df_test.X3.tolist())
tokenizer4 = Tokenizer()
tokenizer4.fit_on_texts(df_train.X4.tolist() + df_test.X4.tolist())
tokenizer5 = Tokenizer()
tokenizer5.fit_on_texts(df_train.X5.tolist() + df_test.X5.tolist())
tokenizer6 = Tokenizer()
tokenizer6.fit_on_texts(df_train.X6.tolist() + df_test.X6.tolist())
tokenizer8 = Tokenizer()
tokenizer8.fit_on_texts(df_train.X8.tolist() + df_test.X8.tolist())
```

```python
y = df_train.y
X = df_train.drop('y', axis=1, inplace=False)
X.X0 = tokenizer0.texts_to_sequences(X.X0)
X.X1 = tokenizer1.texts_to_sequences(X.X1)
X.X2 = tokenizer2.texts_to_sequences(X.X2)
X.X3 = tokenizer3.texts_to_sequences(X.X3)
X.X4 = tokenizer4.texts_to_sequences(X.X4)
X.X5 = tokenizer5.texts_to_sequences(X.X5)
X.X6 = tokenizer6.texts_to_sequences(X.X6)
X.X8 = tokenizer8.texts_to_sequences(X.X8)
X.X0 = X.X0.apply(lambda x: int(x[0]))
X.X1 = X.X1.apply(lambda x: int(x[0]))
X.X2 = X.X2.apply(lambda x: int(x[0]))
X.X3 = X.X3.apply(lambda x: int(x[0]))
X.X4 = X.X4.apply(lambda x: int(x[0]))
X.X5 = X.X5.apply(lambda x: int(x[0]))
X.X6 = X.X6.apply(lambda x: int(x[0]))
X.X8 = X.X8.apply(lambda x: int(x[0]))
X_train,X_valid,y_train,y_valid = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_test = df_test
X_test.X0 = tokenizer0.texts_to_sequences(X_test.X0)
X_test.X1 = tokenizer1.texts_to_sequences(X_test.X1)
X_test.X2 = tokenizer2.texts_to_sequences(X_test.X2)
X_test.X3 = tokenizer3.texts_to_sequences(X_test.X3)
X_test.X4 = tokenizer4.texts_to_sequences(X_test.X4)
X_test.X5 = tokenizer5.texts_to_sequences(X_test.X5)
X_test.X6 = tokenizer6.texts_to_sequences(X_test.X6)
X_test.X8 = tokenizer8.texts_to_sequences(X_test.X8)
X_test.X0 = X_test.X0.apply(lambda x: int(x[0]))
X_test.X1 = X_test.X1.apply(lambda x: int(x[0]))
X_test.X2 = X_test.X2.apply(lambda x: int(x[0]))
X_test.X3 = X_test.X3.apply(lambda x: int(x[0]))
X_test.X4 = X_test.X4.apply(lambda x: int(x[0]))
X_test.X5 = X_test.X5.apply(lambda x: int(x[0]))
X_test.X6 = X_test.X6.apply(lambda x: int(x[0]))
X_test.X8 = X_test.X8.apply(lambda x: int(x[0]))
```

# Training Data

```python
d_train = xgb.DMatrix(X_train, label=y_train)
d_valid = xgb.DMatrix(X_valid, label=y_valid)
watchlist = [(d_train, 'train'), (d_valid, 'valid')]
```

# XGBoost Config and Training

```python
params = {}
params['objective'] = 'multi:softprob'
params['eval_metric'] = 'mlogloss'
params['eta'] = 0.01
# params['num_class'] = 4
params['n_estimators'] = 10000
params['max_depth'] = 4 # or 5
params['min_child_weight'] = 4 # or 1
params['gamma'] = 0.4 # or 0
params['subsample'] = 0.8
params['colsample_bytree'] = 0.75 # or 0.8
params['nthread'] = 4
params['scale_pos_weight'] = 1
params['seed'] = 1
clf = xgb.XGBRegressor()
clf.fit(X, y)
```

# Predict and Write to CSV

```python
y_pred = clf.predict(X_test)
```

```python
submission = pd.DataFrame({'ID': X_test.index.tolist(), 'y': y_pred})
```

```python
submission.head()
```

```python
submission.to_csv('xgb_submission.csv', index=None)
```

# LightGBM Config and Training

```python
params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': {'l2', 'auc'},
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_valid, y_valid, reference=lgb_train)

gbm = lgb.train(params,
                lgb_train,
                num_boost_round=20,
                valid_sets=lgb_eval,
                early_stopping_rounds=5)
```

# Predict and Write to CSV

```python
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

```python
submission = pd.DataFrame({'ID': X_test.index.tolist(), 'y': y_pred})
```

```python
submission.head()
```

```python
submission.to_csv('gbm_submission.csv', index=None)
```
